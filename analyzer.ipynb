### pyt.pytesseract.tesseract_cmd
By default, it is located at: `C:\Program Files\Tesseract-OCR\tesseract.exe`, which is what it I set it to. **Change this if you installed Tesseract-OCR elsewhere.**

```python
import cv2
import pytesseract as pyt
from matplotlib import pyplot as plt
pyt.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

### cv2.imread( ... )
I set this to one of my example screenshots located at: `.\examples\s-normal-1.png`. **Change this if want to use your own screenshot.**

```python
img = cv2.cvtColor(cv2.imread('.\examples\s-normal-1.png')[1230:1355, 180:430, :], cv2.COLOR_BGR2RGB)

#c = r"outputbase digits"
#lang="sdvxDigits"

ret, img = cv2.threshold(img,165,255,cv2.THRESH_BINARY)

plt.imshow(img)
plt.show()
```

## CHIP

```python
chip = img[:, 0:95, :]

cs = pyt.image_to_string(chip[0:30, :, :], lang="sdvxDigits")
cc = pyt.image_to_string(chip[30:60, :, :], lang="sdvxDigits")
cn = pyt.image_to_string(chip[60:90, :, :], lang="sdvxDigits")
ce = pyt.image_to_string(chip[90:120, :, :], lang="sdvxDigits")
print("{}\n{}\n{}\n{}".format(cs, cc, cn, ce))
plt.imshow(chip)
plt.show()
```

## LONG

```python
long = img[:, 80:175, :]
ls = pyt.image_to_string(long[0:30, :, :], lang="sdvxDigits")
lc = pyt.image_to_string(long[30:60, :, :], lang="sdvxDigits")
ln = pyt.image_to_string(long[60:90, :, :], lang="sdvxDigits")
le = pyt.image_to_string(long[90:120, :, :], lang="sdvxDigits")
print("{}\n{}\n{}\n{}".format(ls, lc, ln, le))
plt.imshow(long)
plt.show()
```

## VOL

```python
vol = img[:, 160:250, :]
vs = pyt.image_to_string(vol[0:30, :, :], lang="sdvxDigits")
vc = pyt.image_to_string(vol[30:60, :, :], lang="sdvxDigits")
vn = pyt.image_to_string(vol[60:90, :, :], lang="sdvxDigits")
ve = pyt.image_to_string(vol[90:120, :, :], lang="sdvxDigits")
print("{}\n{}\n{}\n{}".format(vs, vc, vn, ve))
plt.imshow(vol)
plt.show()
```
